# Clean


```python
#copy original datasets 
archive_clean = archive_df.copy()
image_predictions_clean = image_predictions_df.copy()
api_clean =  api_df.copy()
```

## quality
- missing values   : in_reply_to_status_id','retweeted_status_id','retweeted_status_timestamp','retweeted_status_user_id','in_reply_to_user_id'

### Define 
- drop columns (in_reply_to_status_id','retweeted_status_id','retweeted_status_timestamp','retweeted_status_user_id','in_reply_to_user_id')


### Code

```python
archive_clean=archive_clean.query('in_reply_to_status_id == "NaN" &\
                                                     in_reply_to_user_id == "NaN" &\
                                                    retweeted_status_id == "NaN" &\
                                                    retweeted_status_user_id == "NaN"')

drop_columns = ['in_reply_to_status_id','retweeted_status_id',
            'retweeted_status_timestamp','retweeted_status_user_id','in_reply_to_user_id']
archive_clean = archive_clean.drop(columns = drop_columns, axis=1)
```

### Test

```python
archive_clean.info()
```

# Tidness
- The values of four columns (doggo,floofer,pupper,puppo) in twitter_archive dataset should be in one column dog_stage with a category datatype

## Define
- replace the 'None' string with np.nan in the selected columns
- create a dog_stage column and join all the values in the selected values droping nan
- convert the dog_stage column type to categorical
- drop the columns related to the previous 4 stages

## Code

```python
dogs_type = ['doggo', 'floofer', 'pupper', 'puppo']
# replace None values to np.nan
archive_clean[dogs_type] = archive_clean[dogs_type].replace('None', np.nan)

archive_clean['dog_stage'] = archive_clean[dogs_type].\
                                        apply(lambda x: ', '.join(x.dropna().astype(str)),axis =1)
# replace emtpy strings with np .nan
archive_clean.dog_stage = archive_clean.dog_stage.replace('', np.nan).astype('category')
#drop columns
archive_clean = archive_clean.drop(columns = dogs_type, axis =1)

```

# Quality 
- HTML format in source coulmn (<a...)

## Define 
- make a function extract to fix the source column

## Code

```python
# function to extract url text inside tags
def extract(source):
    x= source.find('>') + 1
    y =source[x:].find('<')
    return source[x:][:y]
```

```python
master.source= master.source.apply(lambda x: extract(x))
```
